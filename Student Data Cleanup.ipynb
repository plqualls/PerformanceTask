```python
#rename headers
header_names=['Meets/Tests Rate','Masters/Test Rate','Approaches/Test Rate','Meets/ELA-Reading','Masters/ELA-Reading','Approaches/ELA-Reading','CAMPUSID']
file=pd.read_csv("Working Files/CAMP_D1_STAAR1_final.csv",header=None,skiprows=1,names=header_names)

#print results
file.head()
```

```python
#Load CSV files
achievement = pd.read_csv('Student_Achievement.csv')
rating = pd.read_csv('Campus_Rating.csv')
```
